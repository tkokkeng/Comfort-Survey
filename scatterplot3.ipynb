#### Using matplotlib

```python
from matplotlib.legend import Legend

scale = .05
fig, ax = plt.subplots(figsize=(10, 10))
plots = []
for cat in all_data['Type'].cat.categories:
    plots.append(plt.scatter(x=all_data.loc[all_data['Type'] == cat, 'BTU/Area'], y=all_data.loc[all_data['Type'] == cat, 'PWM/Area'],
                           s=all_data.loc[all_data['Type'] == cat, 'GFA sqm'] * scale, label=cat,
                           alpha=.5, edgecolors='none'))

plt.xlabel('BTU/Area')
plt.ylabel('PWM/Area')
plt.title('PWM, BTU and Building Area')
plt.legend(markerscale=.25)

# Legend for size
dummy = []
for area in [50000, 25000, 500]:
    dummy.append(plt.scatter([], [], c='k', alpha=0.3, s=area*scale, label=str(area)))
leg = Legend(ax, dummy, labels = ['50000', '25000', '500'], scatterpoints=1, frameon=True, labelspacing=2, loc='center right', borderpad=2)
ax.add_artist(leg)
```
